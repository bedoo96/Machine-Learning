#### Prediction using loop

```python
def predict_single_loop(w, x, b):

    n = x.shape[0]
    p = 0

    for i in range(n):
        p_i = x[i] * w[i]
        # print(f"{x[i]} * {w[i]} = {p_i}")
        p += p_i
        # print("running total =", p)

    # print("before bias =", p)
    p += b
    # print("after bias =", p)

    return p
```

```python
# get a row from our training data
x_vec = X_train[0,:] 
print(f"x_vec shape : {x_vec.shape}, x_vec value: {x_vec}")

# make a prediction 
f_wb = predict_single_loop(w_init, x_vec, b_init)
print(f"f_wb shape {f_wb.shape}, prediction : {f_wb}")
```

#### Prediction using dot

```python
def predict(w, x, b):
    """
            single predict using linear regression
    Args:
        x (ndarray): shape(m,n) example with multiple features
        w (ndarray) :shape(n,) model parameter
        b(scalar) : model parameter

    returns:
        p(scalar)
    """
    p = np.dot(x, w) + b
    return p 
```

```python
x_vec = X_train[0,:]
f_wb = predict (w_init, x_vec, b_init)

print(f"f_wb shape {f_wb.shape}, prediction : {f_wb}")
```

### Compute cost with multiple variables

The equation for the cost function  with multiple variables $J(\mathbf{w}, b)$ is:
$$J(\mathbf{w}, b) = \frac{1}{2m} \sum\limits_{i = 0}^{m-1} (f_{\mathbf{w}, b}(\mathbf{x}^{(i)}) -y^{(i)})^2 \tag{1}$$
where:
$$ f_{\mathbf{w}, b}(\mathbf{x}^{(i)}) = \mathbf{w} \cdot \mathbf{x}^{i} + b \tag{2} $$

```python
def compute_cost(x, w, b, y) :
    """
        compute cost.
        Args: 
            X (ndarray (m,n)): data, m example and n features
            y (ndarray(m)) : target values
            w (ndarray (n,)) : model parameter
            b (scalar) : model parameter 

        Return :
        cost (scalar): cost
    """
    m = x.shape[0]
    cost = 0.0
    for i in range(m):
        f_wb_i = np.dot(x[i], w) + b
        cost = cost + (f_wb_i - y[i])**2   
    cost = cost / (2 * m) 
    
    return cost
```

```python
cost = compute_cost(X_train, w_init, b_init, y_train)
print(f"cost at optimal w : {cost}")
```
